# Model 93

```python
plt.rcParams['font.size'] = 35
```

```python
foldernumber = '93'

_,ax = plt.subplots(figsize=(15,10))
low_lim = 0
up_lim = 1500
alpha = 1.
lw = 5

colors = ['blue','black','red']
keys = ['pe','force','stress']
datafolder = 'loss_weights_data/'

for f in os.listdir(datafolder):
    if f.startswith(foldernumber):
        foldername = os.path.join(datafolder,f)

for i,key in enumerate(keys):
    x = np.arange(len(np.loadtxt(os.path.join(foldername,f'{key}.txt'))))[low_lim:up_lim]
    ax.plot(x, np.loadtxt(os.path.join(foldername,f'{key}.txt'))[low_lim:up_lim], '--', lw=lw, alpha=alpha, c=colors[i], label=key)

ax.set_xlabel(r'Epoch $i$')
ax.set_ylabel(r'Prefactor value $\alpha_k^{(i)}$')
# ax.set_xlim([0., 250])
ax.set_ylim([0.0, 1.])
ax.legend()
ax.grid()
plt.savefig(f'loss_weights_figures/loss_weights_{foldernumber}_1500epochs.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
print('last values')
for key in keys:
    last_value = cumulative_moving_average(np.loadtxt(os.path.join(foldername,f'{key}.txt')))[-1]
    print(key, f'{last_value:0.4f}')
```
